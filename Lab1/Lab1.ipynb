#  Common code

```python
def calculate_mean_word_length(input_text):
    words = input_text.split()
    words_length = [len(word) for word in words]
    return sum(words_length)/len(words)
```

```python
def generate(generator,character_set,lenght,prob_distribution = None):
    generator_output = generator(character_set,lenght,prob_distribution)
    return generator_output
```

```python
from collections import Counter, OrderedDict, deque
```

```python
def character_frequency(text):
    counter_dict = dict(sorted(Counter(text).items(),key=lambda x:x[1]))
    character_counter = sum(counter_dict.values())
    return {key:value/character_counter for key,value in counter_dict.items()}
```

```python
def compute_ngram_probabilty(text_corpus,ngram_size):
        ngrams = []
        for i,_ in enumerate(text_corpus[:-ngram_size]):
            ngram = text_corpus[i:i+ngram_size]
            ngrams.append(ngram)
        else:
            #Last iteration is executed in secial way. iterable[:-0] return []
            ngram = text_corpus[-ngram_size:]
            ngrams.append(ngram)
        counter_dict = Counter(ngrams)

        omega = len(ngrams) + ngram_size - 1
        prob_dict = {key:value/omega for key, value in counter_dict.items()}
        return prob_dict
```

```python
bigrams_prob = compute_ngram_probabilty(full_text,2)
```

```python
# This probability isn't equeal 1.0 but it's really close. 
# The error is due to the distortion of values on floating point operations
# For the purposes of this exercise, I will assume that this error is insignificant.

# After Lab_1 pdf update
# Additonal factor is that now I ignore probability for n-1 n-grams at the end of text.
# Alpha contains three 3-grams ['Alp','lph','pha']. I don't include ['ha ','a  ']
# This error is insignificant if big text corpus is used.
```

```python
def compute_probabilty_afer_character(character, bigrams_prob,unigram_prob):
    result_dict = {}
    first_char_prob = unigram_prob[character]
    for second_char in unigram_prob.keys():
        result_key = character+second_char
        try:
            result_dict[result_key] = bigrams_prob[result_key]/first_char_prob
        except KeyError:
            result_dict[result_key] = '0.0'
    return result_dict
```

# Prawdopodobieństwo znaków po zanku 'e'

```python
e_result = compute_probabilty_afer_character('e',bigrams_prob,unigrams_prob)
for key in sorted (e_result.keys(),key=lambda x:x[1]) :  
    print(f'{key} {e_result[key]}')
```

```python
# This fucntion can be used to include prob of engrams at the end of text, by replace last iteration code.
# If you use this function make sure that remove ' + ngram_size -1' from omega computation and 
# extend list of ngrams by result of this function.
def alterative_last_iteration_including_end_of_file(text,ngram_size):
    result = []
    for i in range(ngram_size,0,-1):
        part_1 = text[-i:]
        part_2 = ''.join([' ']*(ngram_size+1 - i))
        result.append(part_1 + part_2)
    return result
```

```python
class MarkovSource:
    
    def __init__(self,source_level,text_corpus):
        self.text_corpus = text_corpus
        self.source_level = source_level
        self.prob = self._compute_probabilty(self.text_corpus,self.source_level+1)
        self.lower_level_source_instance = None
        
    def _compute_probability_letter_after_ngram(self,ngrams_with_letter,ngram_size): 
        temp_name = {}
        for ngram_with_letter in ngrams_with_letter:
            ngram = str(ngram_with_letter[:-1])
            letter = str(ngram_with_letter[-1])
            if ngram in temp_name:
                temp_name[ngram].append(letter)
            else:
                temp_name[ngram] = [letter]
                
        omega = len(ngrams_with_letter) + ngram_size -1  
        result = {}       
        for ngram,letters in temp_name.items():
            counter_dict = Counter(letters)
            prob_dict = {key:value/omega for key, value in counter_dict.items()}
            result[ngram] = prob_dict
        return result
    
    def _compute_probabilty(self,text_corpus,ngram_size):
        ngrams_with_letter = []
        
        for i,_ in enumerate(text_corpus[:-ngram_size]):
            ngram_with_letter = text_corpus[i:i+ngram_size]
            ngrams_with_letter.append(ngram_with_letter)
        else:
            #Last iteration is executed in sepcial way. iterable[:-0] return []
            ngram_with_letter = text_corpus[-ngram_size:]
            ngrams_with_letter.append(ngram_with_letter)
            #ngrams_with_letter.extend(alterative_last_iteration_including_end_of_file(text_corpus,ngram_size))
        return self._compute_probability_letter_after_ngram(ngrams_with_letter,ngram_size)   
    
    
    
    def generate_next(self,input_state):
        # If len of input state is inappropriate then try use lower level source or rise exception.
        if len(input_state)>=self.source_level:
            if len(input_state)>self.source_level:
                input_state = input_state[-self.source_level:]
            state_dict = self.prob[input_state]
            return random.choices(list(state_dict.keys()),list(state_dict.values()))[0]
        else:
            if self.lower_level_source_instance is None:
                self.lower_level_source_instance = MarkovSource(self.source_level - 1,self.text_corpus)
            return self.lower_level_source_instance.generate_next(input_state)
```

```python
def generate_zad_5(markov_source,length,input_state=''):
    queue = deque(list(input_state),maxlen = markov_source.source_level)
    output = []
    for letter in input_state:
        queue.append(letter)
        output.append(letter)
        
    for i in range(length):
        output_char = markov_source.generate_next("".join(queue))
        queue.append(output_char)
        output.append(output_char)
    return "".join(output)
```
